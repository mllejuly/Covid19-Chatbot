```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Learning Rate

```python
lr=["2e-5","2e-7","2e-9","2e-10","2e-15"]

# distilbert-base-nli-stsb-mean-tokens, eps=3, batch_size=16
acc1=[0.2857, 0.4286, 0.4286, 0.4286, 0.4286]

# bert-base-nli-mean-tokens, eps=3, batch_size=16
acc2=[0.3333, 0.5238, 0.5714, 0.5714, 0.5714]

line1=plt.plot(lr,acc1,color='#33A6BD',label='distilbert-base-nli-stsb-mean-tokens')
line2=plt.plot(lr,acc2,color='#FE795C',label='bert-base-nli-mean-tokens')

plt.plot(lr,acc1,color='#33A6BD',marker='^',markersize=10)
plt.plot(lr,acc2,color='#FE795C',marker='o',markersize=8)

plt.title('Learning Rate Impact on Model Accuracy')
plt.xlabel('Learning Rate')
plt.ylabel('Accuracy')
plt.legend()
plt.savefig('Learning_Rate_Impact_on_Model_Accuracy.png')
plt.show()
```

## Loss

```python
def plot_3d(df, model_type, col_x, col_y ,col_z, col_size, title):
    fig = plt.figure(figsize=(20, 15))
    ax = fig.add_subplot(111, projection='3d')

    df = df[df.base_model == model_type]
    xs = df.loc[:, col_x]
    ys = df.loc[:, col_y]
    zs = df.loc[:, col_z]
    ss = df.loc[:, col_size]*1000
    ax.scatter(xs, ys, zs, s=ss)

    ax.set_xlabel(col_x, fontsize=25)
    ax.set_ylabel(col_y, fontsize=25)
    ax.set_zlabel(col_z, fontsize=25)
    plt.title(title, fontsize=30)

    plt.show()

df_perf = pd.read_csv('loss/loss_analysis.csv')
```

```python
model = 'bert-base-nli-mean-tokens'
plot_3d(df_perf, model,
        'batch_size', 'epoch', 'learning_rate', 'loss',
        'The Scatter Plot of the Loss Trend\n{}'.format(model))
```

```python
model = 'distilbert-base-nli-stsb-mean-tokens'
plot_3d(df_perf, model,
        'batch_size', 'epoch', 'learning_rate', 'loss',
        'The Scatter Plot of the Loss Trend\n{}'.format(model))
```
